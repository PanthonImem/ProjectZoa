```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('dataset.csv')
df.head(400)
```

```python
num = df['NumZoa'].values
numfoo = df['NumFood'].values
Velo = df['Velo'].values
MaxHealth = df['MaxHealth'].values
Sight = df['Sight'].values
Gen = df['Gen'].values
Size = df['Size'].values
NumEaten = df['NumEaten'].values
ls = []
ls2 = []
afactor = sum(numfoo)/sum(num)
for i in range(len(numfoo)):
    ls.append(numfoo[i]/afactor)
    ls2.append(1)
def adjust(l):
    adj = sum(l)
    ls = []
    for member in l:
        ls.append( member*100/adj)
    return ls
    
minVelo = df['minVelo'].values
minMaxHealth = df['minMaxHealth'].values
minSight = df['minSight'].values
minGen = df['minGen'].values
minSize = df['minSize'].values
minNumEaten = df['minNumEaten'].values

maxVelo = df['maxVelo'].values
maxMaxHealth = df['maxMaxHealth'].values
maxSight = df['maxSight'].values
maxGen = df['maxGen'].values
maxSize = df['maxSize'].values
maxNumEaten = df['maxNumEaten'].values
```

```python
plt.plot(np.arange(len(num)), numfoo , color='green')
plt.xlabel('Iter')
plt.ylabel('NumFood')
plt.title('Number of Food')
#plt.savefig('../Graph')
plt.show()
plt.plot(np.arange(len(num)), num , color='red')
plt.xlabel('Iter')
plt.ylabel('NumZoa')
plt.title('Number of Zoa')
#plt.savefig('../Graph')
plt.show()

plt.plot(np.arange(len(num)), num , color='red')
plt.plot(np.arange(len(num)), ls , color='green')
plt.plot(np.arange(len(num)), ls2 , color='gold')
plt.xlabel('Iter')
plt.ylabel('Num Zoa & num Food')
plt.title('Number of Zoa and Number of Food')
#plt.savefig('../Graph')
plt.show()
plt.plot(np.arange(len(num)), Velo , color='blue')
plt.plot(np.arange(len(num)), minVelo , color='red')
plt.plot(np.arange(len(num)), maxVelo , color='purple')
plt.xlabel('Iter')
plt.ylabel('Average Velocity')
plt.title('Velo_and_iter')
#plt.savefig('../Graph')
plt.show()

plt.plot(np.arange(len(num)), MaxHealth , color='blue')
plt.plot(np.arange(len(num)), minMaxHealth , color='red')
plt.plot(np.arange(len(num)), maxMaxHealth , color='purple')
plt.xlabel('Iter')
plt.ylabel('Average Health')
plt.title('Health_and_iter')
#plt.savefig('../Graph')
plt.show()
plt.plot(np.arange(len(num)), Sight , color='blue')
plt.plot(np.arange(len(num)), minSight , color='red')
plt.plot(np.arange(len(num)), maxSight , color='purple')
plt.xlabel('Iter')
plt.ylabel('Average Sight')
plt.title('Sight_and_iter')
#plt.savefig('../Graph')
plt.show()
plt.plot(np.arange(len(num)), Gen , color='blue')
plt.plot(np.arange(len(num)), minGen , color='red')
plt.plot(np.arange(len(num)), maxGen , color='purple')
plt.xlabel('Iter')
plt.ylabel('Average Generation')
plt.title('Gen_and_iter')
#plt.savefig('../Graph')
plt.show()
#plt.plot(np.arange(len(num)), num , color='gold')
plt.plot(np.arange(len(num)), Size , color='blue')
plt.plot(np.arange(len(num)), minSize , color='red')
plt.plot(np.arange(len(num)), maxSize , color='purple')
plt.xlabel('Iter')
plt.ylabel('Average Size')
plt.title('Size_and_iter')
#plt.savefig('../Graph')
plt.show()

plt.plot(np.arange(len(num)), NumEaten , color='blue')
plt.plot(np.arange(len(num)), minNumEaten , color='red')
plt.plot(np.arange(len(num)), maxNumEaten , color='purple')
plt.xlabel('Iter')
plt.ylabel('Average Num Eaten')
plt.title('NumEaten_and_iter')
#plt.savefig('../Graph')
plt.show()

plt.plot(np.arange(len(num)), adjust(Velo) , color='purple')
plt.plot(np.arange(len(num)), adjust(MaxHealth) , color='green')
plt.plot(np.arange(len(num)), adjust(Sight) , color='gold')
#plt.plot(np.arange(len(num)), adjust(Gen) , color='red')
plt.plot(np.arange(len(num)), adjust(Size) , color='brown')
plt.xlabel('Iter')
plt.ylabel('??')
plt.title('Everything')
#plt.savefig('../Graph')
plt.show()
```
